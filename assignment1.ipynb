```python
#Task 1: Data Preparation
# "You will start by loading the CSV data from the file (using appropriate pandas functions) and checking whether the loaded data is equivalent to the data in the source CSV file.
# Then, you need to clean the data by using the knowledge we taught in the lectures. You need to deal with all the potential issues/errors in the data appropriately (such as: typos, extra whitespaces, sanity checks for impossible values, and missing values etc). "

# Please structure code as follows: 
# always provide one line of comments to explain the purpose of the code, e.g. load the data, checking the equivalent to original data, checking typos (do this for each other types of errors)


#load the data
import pandas as pd
import numpy as np 

df = pd.read_csv("/Users/admin/Desktop/Automobile.csv", sep='#', decimal='.', header=None, names=['symboling', 'normalized-losses', 'make', 'fuel-type','aspiration', 'num-of-doors','body-style', 'drive-wheels', 'engine-location','wheel-base', 'length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price'] )

#checking shape and column names to match with original data 
df.shape 



#removing extra white space, typos and correcting case mismatch in make attribute
df['make'] = df['make'].str.strip()
df['make'] = df['make'].str.lower()
df['make'].replace(['vol00112ov'], ['volvo'], inplace = True)
df['make'].replace(['alfa-romero'], ['Alfa-romeo'], inplace = True)
df['make'].replace(['peugot'],['Peugeot'], inplace = True)

 
#removing extra white space, correcting case mismatch in fuel-type attribute
df['fuel-type'] = df['fuel-type'].str.strip()
df['fuel-type'] = df['fuel-type'].str.lower()

#removing extra white space, typos and correcting case mismatch in aspiration attribute
df['aspiration'] = df['aspiration'].str.strip()
df['aspiration'] = df['aspiration'].str.lower()
df['aspiration'].replace(['turrrrbo'], ['turbo'], inplace = True)

#removing extra white space, typos and correcting case mismatch in num-of-doors attribute
df['num-of-doors'] = df['num-of-doors'].str.strip()
df['num-of-doors'].replace(['fourr'], ['four'], inplace = True)
df['num-of-doors'] = df['num-of-doors'].str.lower()

 
#removing extra white space, correcting case mismatch in body-style attribute
df['body-style'] = df['body-style'].str.strip()
df['body-style'] = df['body-style'].str.lower()

 
#removing extra white space, correcting case mismatch in drive-wheels attribute
df['drive-wheels'] = df['drive-wheels'].str.strip()
df['drive-wheels'] = df['drive-wheels'].str.lower()

#removing extra white space, correcting case mismatch in engine-location attribute
df['engine-location'] = df['engine-location'].str.strip()
df['engine-location'] = df['engine-location'].str.lower()

#removing extra white space, correcting case mismatch in engine-location attribute
df['num-of-cylinders'] = df['num-of-cylinders'].str.strip()
df['num-of-cylinders'] = df['num-of-cylinders'].str.lower()


#inserting mean for missing values in the following attributes horsepower,peak-rpm,bore,stroke,price,normalized-losses,
#num-of-doors

df['horsepower'].fillna(df['horsepower'].mean(axis=0),inplace=True)

df['peak-rpm'].fillna(df['peak-rpm'].mean(axis=0),inplace=True)

df['bore'].fillna(df['bore'].mean(axis=0),inplace=True)

df['stroke'].fillna(df['stroke'].mean(axis=0),inplace=True)

df['price'].fillna(df['price'].mean(axis=0),inplace=True)

df['normalized-losses'].fillna(df['normalized-losses'].mean(axis=0),inplace=True)

#inserting  mode of num-of-door in place of missing values  because it is nominal value

df['num-of-doors'].mode()

df['num-of-doors'] = df['num-of-doors'].replace(np.nan, 'four')


#removing impossible values 
df.drop(df.loc[df['symboling']==4].index, inplace=True)

df.drop(df.loc[df['price']== 0.000000 ].index, inplace=True)


#changing datatype  of ordinal values and categorize them

df['num-of-cylinders'] = pd.Categorical(df['num-of-cylinders'], categories = ['two',   'three','four' , 'five','six',  'eight', 'twelve'], ordered = True)


df['symboling'] = pd.Categorical(df['symboling'], categories = [-2,   -1,0 ,1,2,  3], ordered = True)



```

```python
# Task 2: Data Exploration
# 1. Choose 1 column with nominal values, 1 column with ordinal Values, and 1 column with numerical values


#Code goes after this line

import matplotlib.pyplot as plt




df['symboling'] = pd.Categorical(df['symboling'], categories = [-2,   -1,0 ,1,2,  3], ordered = True)



df.groupby('symboling')['symboling'].count().plot(figsize=(10,8),kind='bar',color='purple')

plt.title("insurance risk rating")
plt.show()




df['horsepower'].plot(kind='hist',bins =6 ,color='green')

plt.title("horsepower")
plt.show()



df['num-of-doors'].value_counts().plot(kind='pie',autopct='%.2f')
plt.title("num-of-doors")
plt.show()


```
